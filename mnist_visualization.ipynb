```python
from sklearn.datasets import load_digits
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
from sklearn.preprocessing import StandardScaler
import seaborn as sns
%matplotlib inline
```

Load the data and targets

```python
data = load_digits().data 
targets = load_digits().target
```

Initialize the functions

```python
n_components=3
pca = PCA(n_components=n_components)
tsne = TSNE(n_components=n_components)
scaler = StandardScaler()
```

Scale data either between $[0, 1]$, i.e. `data/255` or with scaler 

```python
# transformed_data = scaler.fit_transform(data)
transformed_data = data / 255.
print(transformed_data)
```

Do the dimensionality reduction

```python
pca_transformed = pca.fit_transform(transformed_data)
```

```python
sns.scatterplot(pca_transformed[:, 0], pca_transformed[:, 1], palette=sns.color_palette("hls", 10), hue=targets, alpha=0.5)
```
